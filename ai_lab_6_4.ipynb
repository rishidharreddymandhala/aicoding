```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    # Method to display student details including name, roll number and marks
    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Marks: {self.marks}")

    # Method to check whether the student's marks are above the class average
    # Use if-else condition and return an appropriate performance message
    def check_performance(self, class_average):
        if self.marks >= class_average:
            return f"{self.name} is performing above average."
        else:
            return f"{self.name} is performing below average."

# Create a Student object with sample values
student_obj = Student("John Doe", "A001", 85)

# Call the method to display student details
student_obj.display_details()

# Define a class average marks value
class_avg = 75

# Call the performance checking method and print the returned message
performance_status = student_obj.check_performance(class_avg)
print(performance_status)
```

```python
sensor_readings = [3, 6, 9, 12, 15, 20]

# iterate over each sensor reading
for reading in sensor_readings:
    # check if the reading is an even number using modulus operator
    if reading % 2 == 0:
        # if even, calculate the square of the number
        square = reading ** 2
        # print the reading and its square in a readable format
        print(f"Reading: {reading}, Square: {square}")
```

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    # Method to deposit money into the account
    def deposit(self, amount):
        self.balance += amount
        print(f"Successfully deposited ${amount}. New balance: ${self.balance}")

    # Method to withdraw money from the account
    def withdraw(self, amount):
        # Check if sufficient balance is available
        if amount <= self.balance:
            self.balance -= amount
            print(f"Successfully withdrew ${amount}. Remaining balance: ${self.balance}")
        else:
            print(f"Insufficient funds. Withdrawal of ${amount} failed. Current balance: ${self.balance}")

# Create a BankAccount object with sample values
my_account = BankAccount("Alice Smith", 500)

# Perform a deposit transaction
my_account.deposit(200)

# Perform a withdrawal transaction with sufficient balance
my_account.withdraw(150)

# Perform a withdrawal transaction with insufficient balance
my_account.withdraw(1000)
```

```python
# list of students with their names and scores
students = [
    {"name": "Rishi", "score": 82},
    {"name": "Ananya", "score": 74},
    {"name": "Kiran", "score": 90},
    {"name": "Suman", "score": 68},
    {"name": "Priya", "score": 88}
]

# initialize index for while loop
index = 0

# iterate through the students list using while loop
while index < len(students):
    # check if the student's score is greater than 75
    if students[index]["score"] > 75:
        # if eligible, print the student's name in a clean format
        print(f"Eligible Student: {students[index]['name']}")

    # increment the index properly to avoid infinite loop
    index += 1
```

```python
class ShoppingCart:
    def __init__(self):
        # Initialize an empty list to store items as dictionaries
        self.items = []

    # Method to add an item to the shopping cart
    def add_item(self, name, price, quantity):
        item = {
            "name": name,
            "price": price,
            "quantity": quantity
        }
        self.items.append(item)
        print(f"Added {quantity} x {name} to the cart.")

    # Method to remove an item from the cart by its name
    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                print(f"Removed {name} from the cart.")
                return
        print(f"Item '{name}' not found in the cart.")

    # Method to calculate the total bill using a loop
    def calculate_total(self):
        total = 0
        for item in self.items:
            total += item["price"] * item["quantity"]
        return total

    # Method to apply a conditional discount if the total exceeds a threshold
    def apply_discount(self, threshold, discount_percent):
        current_total = self.calculate_total()
        if current_total > threshold:
            discount_amount = (discount_percent / 100) * current_total
            final_total = current_total - discount_amount
            print(f"Total: ${current_total:.2f}. Threshold reached! Applied {discount_percent}% discount.")
            return final_total
        else:
            print(f"Total: ${current_total:.2f}. No discount applied (threshold of ${threshold} not met).")
            return current_total

    # Helper method to display the items in the cart
    def display_cart(self):
        print("\n--- Shopping Cart Items ---")
        for item in self.items:
            print(f"{item['name']} - Price: ${item['price']}, Quantity: {item['quantity']}")
        print("---------------------------\n")

# --- Sample Input/Output ---

# Create a ShoppingCart object
my_cart = ShoppingCart()

# Add items
my_cart.add_item("Laptop", 1200, 1)
my_cart.add_item("Mouse", 25, 2)
my_cart.add_item("Keyboard", 45, 1)

# Display current cart
my_cart.display_cart()

# Calculate total before discount
total_before = my_cart.calculate_total()
print(f"Current Total: ${total_before:.2f}")

# Remove an item
my_cart.remove_item("Mouse")

# Apply discount (e.g., 10% off if total > $1000)
final_bill = my_cart.apply_discount(threshold=1000, discount_percent=10)
print(f"Final Bill to pay: ${final_bill:.2f}")

# Try applying discount with a higher threshold that isn't met
final_bill_no_discount = my_cart.apply_discount(threshold=5000, discount_percent=20)
print(f"Final Bill (no discount scenario): ${final_bill_no_discount:.2f}")
```
